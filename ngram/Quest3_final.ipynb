```python
# 파일 불러오기 
with open('/content/drive/MyDrive/Colab Notebooks/06TheAvengers.txt','r') as f:
  data = f.readlines()

print(len(data))
```

```python
data[:-10]
```

```python
# Counter 임포트 하기
from collections import Counter
import string
import re
```

```python
# 특수 문자 제거 (창호 퍼실님의 도움을 받음.)
word_list = [] 

switch = ''
for i in range(len(string.punctuation)):
    switch = switch + ' '

table = str.maketrans(string.punctuation, switch)

for sen in data:
    sen = sen.replace('\n', ' ')
    sen = sen.lower()
    sen = sen.translate(table)
    splited_sen = sen.split()
    word_list += list(zip(splited_sen, splited_sen[1:]))

# 실패한 코드 크게 달라 보이지 않지만 , switch 변수의 역활이 크다.
# word_list 를 만드는 과정이 상당히 단순명료 하다. 

# sen_list = []
# for sen in data:
#   sen1 = sen.replace("\n"," ")
#   sen1 = sen.translate(str.maketrans(' ',' ',string.punctuation))
#   sen2 = sen1.lower()
#   sen_list.append(sen2)
word_list[:30]
```

```python
# # 단어 단위 split
# word_list=[]
# for word in sen_list:
#   word1 = word.split()
#   for i in range(len(word1) - 1):
#     gram_temp = (word1[i], word1[i + 1])
#     word_list.append(gram_temp)
# word_list[:10]
```

```python
# 전체 2gram dictinary
dict_2gram = dict(Counter(word_list))
max2gram = max(dict_2gram,key = dict_2gram.get)
max2gram_word = max(dict_2gram.values())
```

```python
# 가장 많이 나오는 2gram 과 빈도수
print(max2gram , max2gram_word)
```

```python
# 상위 20 개의 2gram 단어
top20_2gram = dict(Counter(word_list).most_common(20))
print(top20_2gram)
```
